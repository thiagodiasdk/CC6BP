```python
#This function writes the minors only in terms of mutual distances
def computedetnewtnian(II):
    list=[]
    #Define the variables of the problem
    var('S125,S135,S145,S165,S215,S235,S245,S265,S315,S325,S345,S365,S415,S425,S435,S465,R12,R13,R14,R15,R23,R24,R25,R34,R35,R45,R56,M1,M2,M3,M4,M5')
    
    for k in xrange(1,len(II)+1):
        f=II[k].sage()
        #Eliminating R_{ijk}
        a=f.substitute(S125=R12^(-3)-R25^(-3),S135=R13^(-3)-R35^(-3),S145=R14^(-3)-R45^(-3),S215=R12^(-3)-R15^(-3),S235=R23^(-3)-R35^(-3),S245=R24^(-3)-R45^(-3),
        S315=R13^(-3)-R15^(-3),S325=R23^(-3)-R25^(-3),S345=R34^(-3)-R45^(-3),S415=R14^(-3)-R15^(-3),S425=R24^(-3)-R25^(-3),S435=R34^(-3)-R35^(-3),
        S165=R15^(-3)-R56^(-3),S265=R25^(-3)-R56^(-3),S365=R35^(-3)-R56^(-3),S465=R45^(-3)-R56^(-3))
        #Clearing the denominators
        n=a.numerator()
        D=n
        #colect the rewritten minors
        list.append(D)
    return list        
```

```python
# This function colect the leading monomial of a set of generators of a ideal
def colectleadingterms(I):
    list=[]
    for i in range(1,len(I)+1):
        f=I[i].leadmonom()
        list.append(f)
    return singular.ideal(list)
```
